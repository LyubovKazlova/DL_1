1. Создайте случайный тензор размерности (7,7):

```python
x_tensor = torch.rand(7,7)
describe (x_tensor)
```

2. Определите операцию матричного умножения на тензоре с шага 1 с любым случайным тензором с размерностью (1, 7):

```python
y = torch.rand(1,7)
z = x_tensor @ y.T #use transparent for correct size
describe (z)
```

```python

#something about commutativity

y @ x_tensor
```

3. Найдите максимум, минимум и среднее значение для полученного тензора (шаг 2):

```python
print ("maximum:{},\nminimum: {},\nmean: {}".
       format (torch.max(z), torch.min(z), torch.mean(z)))
```

4. Найдите индекс максимального и минимального значения для тензора, полученного в шаге 2:

```python
print ("index of maximum:{},\nindex of minimum :{}".
      format (torch.argmax(z), torch.argmin(z)))
```

5. Найдите обратную матрицу (тензор), для матрицы (тензора) полученного на шаге 1:

```python
x_inv = torch.linalg.inv(x_tensor)
describe (x_inv)
```

```python
#check our result, for result above 0 right
torch.dist(x_tensor @ x_inv, torch.eye(7))
```
