```python
import torch
```

```python
# function with information about tensor
def describe(x):
    print("Type: {}".format(x.type()))
    print("Shape/Size: {}".format(x.shape))
    print("Values: \n{}".format(x))
```

3. Создайте случайный тензор формы 5 на 3 в интервале [3, 7):

```python

# usually dtype is torch.float64, but in this case it's too much
y = torch.randint(3,7,(5,3), dtype = torch.float32)
describe (y)
```

7. Верните пакетное произведение 2 трехмерных матриц (a = torch.rand(3,4,5) и b = torch.rand(3, 5, 4)):

```python
a = torch.rand(3,4,5)
b = torch.rand(3,5,4)
C = torch.matmul(a,b)
describe (C)
```
